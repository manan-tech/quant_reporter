## 1. Choose a portfolio

Edit `CHOICE` to one of the presets, or drop your own `{ticker: weight}` dict into `MY_PORTFOLIO`.

```python
PRESETS = {
    "60/40":          {"SPY": 0.60, "AGG": 0.40},
    "all-weather":    {"VTI": 0.30, "TLT": 0.40, "IEI": 0.15, "GLD": 0.075, "DBC": 0.075},
    "magnificent-7":  {t: 1/7 for t in ("AAPL","MSFT","GOOGL","AMZN","NVDA","META","TSLA")},
    "bogleheads-3":   {"VTI": 0.60, "VXUS": 0.20, "BND": 0.20},
}

CHOICE = "all-weather"        # <- change me
MY_PORTFOLIO = None            # <- or set e.g. {"AAPL": 0.5, "KO": 0.3, "GLD": 0.2}

portfolio = MY_PORTFOLIO or PRESETS[CHOICE]
BENCHMARK = "SPY"
TRAIN_START, TRAIN_END = "2015-01-01", "2023-12-31"
portfolio
```

## 2. Generate the report

```python
import quant_reporter as qr

qr.create_combined_report(
    portfolio, BENCHMARK, TRAIN_START, TRAIN_END,
    filename="report.html",
    num_simulations=2000,   # lower = faster in Colab
)
print("Report written to report.html")
```
